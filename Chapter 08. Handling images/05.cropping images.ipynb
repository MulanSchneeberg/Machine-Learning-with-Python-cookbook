You want to remove the outer portion of the image to change its dimensions

The image is encoded as a two-dimensional NumPy array, so we can crop the
image easily by slicing the array

```python
# Load image
import cv2
import numpy as np
from matplotlib import pyplot as plt
# Load image in grayscale
image = cv2.imread("../sim_data/images/plane_256x256.jpg", cv2.IMREAD_GRAYSCALE)
# Select first half of the columns and all rows
image_cropped = image[:,:128]
# Show image
plt.imshow(image_cropped, cmap="gray"), plt.axis("off")
plt.show()
```

```python
image_cropped = image[:,128:]
# Show image
plt.imshow(image_cropped, cmap="gray"), plt.axis("off")
plt.show()
```

Since OpenCV represents images as a matrix of elements, by selecting the rows
and columns we want to keep we are able to easily crop the image. Cropping can
be particularly useful if we know that we only want to keep a certain part of
every image. For example, if our images come from a stationary security camera
we can crop all the images so they only contain the area of interest
